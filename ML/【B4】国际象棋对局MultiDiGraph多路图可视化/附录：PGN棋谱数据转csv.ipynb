原始棋局对局信息存储在`PGN`文件中

PGN file (PGN =”Portable Game Notation”)

包含了1886年至1985年，国际象棋锦标赛的685场对局。

原始数据下载：http://chessproblem.my-free-games.com/chess/games/Download-PGN.php

```python
import pandas as pd
```

## 从PGN文件中读取棋局信息

```python
df = pd.DataFrame()
json_temp = {}
with open('WCC.pgn','r') as datafile:
    lines = (line.rstrip('\r\n') for line in datafile)
    for line in lines:
        if line.startswith('['):
            tag, value = line[1:-1].split(' ', 1)
            value = eval(value)
            if tag == 'Event':
                json_temp['Event'] = value
            if tag == 'Site':
                json_temp['Site'] = value  
            if tag == 'Date':
                json_temp['Date'] = value
            if tag == 'Round':
                json_temp['Round'] = value
            if tag == 'White':
                json_temp['White'] = value
            if tag == 'Black':
                json_temp['Black'] = value
            if tag == 'Result':
                json_temp['Result'] = value
            if tag == 'ECO':
                json_temp['ECO'] = value
            if tag == 'EventDate':
                json_temp['EventDate'] = value
                df = df.append(json_temp, ignore_index=True)
                json_temp = {}
```

```python
df
```

## 更改列顺序

```python
columns = ['Date', 'EventDate', 'Event', 'Site', 'ECO', 'White', 'Black', 'Round', 'Result']
df = df[columns]
```

## 导出为csv文件

```python
df.to_csv('WCC.csv', index=False)
```
